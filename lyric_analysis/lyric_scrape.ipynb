```python
def get_url_AZ(artist,song):
    
    song = re.sub('['+string.punctuation+']', '', song).replace(' ','').lower()
    base = 'http://www.azlyrics.com/lyrics/'
    if artist=='dylan':
        artist1='bobdylan'
    elif artist=='jackson':
        artist1='michaeljackson'
    elif artist=='knowles':
        artist1='beyonceknowles'
    elif artist=='spears':
        artist1='britneyspears'   
    elif artist=="lavigne":
        artist1=str('avrillavigne')
    else:
        artist1=str(artist)
    rest = artist1 + '/' + song + '.html'
    return base + rest
```

```python
def get_lyric_AZ(url, user_agents):
    
    r = urllib.urlopen(url).read()

    split = r.split('<!-- Usage of azlyrics.com content by any third-party lyrics provider is prohibited by our licensing agreement. Sorry about that. -->',1)
    
    if len(split)==1:
        return ''
    split_html=split[1]
    split = split_html.split('</div>',1)
    lyrics = split[0];lyrics
    lyrics = re.sub('(<.*?>)'," ",lyrics);lyrics
    lyrics = re.sub('\n'," ",lyrics);lyrics
    lyrics = re.sub('\r'," ",lyrics);lyrics
    sleep(random.randint(10,20))
    return lyrics
    
```

```python
def combine_pickle(artists):
    master_dict={}
    for one_artist in artists:
        one_artist_dict={}
        with open(one_artist +'_song_list.pickle', 'rb' ) as f:
            art_info=pickle.load(f)        
  
        for i in xrange(1, len(art_info[one_artist])+1):
            with open(one_artist + '_lyrics.pickle'+ str(i), 'rb' ) as f:
                one_lyric=pickle.load(f)  
            one_artist_dict.update(one_lyric)
            master_dict.update(one_lyric)
            print i
        pickle.dump(one_artist_dict, open(one_artist + '_lyrics.pickle', 'wb'))
        
    pickle.dump(master_dict, open('all_lyrics.pickle'+ time.strftime("%x"), 'wb'))
    return master_dict
```
